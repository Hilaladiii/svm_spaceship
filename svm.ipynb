```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
import numpy as np
from sklearn.impute import SimpleImputer


categorical_columns = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
continous_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

for col in categorical_columns:
    df[col] = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent').fit_transform(df[[col]].values.reshape(-1,1))[:,0]

for col in continous_columns:
    df[col] = SimpleImputer(missing_values = np.nan, strategy = 'mean').fit_transform(df[[col]].values.reshape(-1,1))[:,0]
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
print(df["HomePlanet"].unique())
print(df["Destination"].unique())
```

```python
df.head()
```

```python
nan_values = df.isna().sum()
print(nan_values)
```

```python
df['Deck'] = df['Cabin'].apply(lambda x: str(x).split('/')[0])
df['Number'] = df['Cabin'].apply(lambda x:  str(x).split('/')[1]).astype(int)
df['Side'] = df['Cabin'].apply(lambda x:  str(x).split('/')[2])

categorical_columns.append('Deck')
categorical_columns.append('Side')
continous_columns.append('Number')
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, MinMaxScaler
encoder = OneHotEncoder()
category_onehot = ["HomePlanet", "Destination"]
for column in category_onehot:
    categorical = pd.DataFrame(df[column])
    transformed = encoder.fit_transform(categorical)
    transformed_df = pd.DataFrame(transformed.toarray(), columns=encoder.get_feature_names_out([column]))
    df = pd.concat([df, transformed_df], axis=1)
    df.drop(columns=[column], inplace=True)

```

```python
categorical_columns.remove('HomePlanet')
categorical_columns.remove('Destination')
```

```python
categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

for col in continous_columns:
        df[col] = np.log1p(df[col])
```

```python
df.head()
```

```python
df.info()
```

```python
le2 = LabelEncoder()
le2.fit(df['Transported'])
y = le2.transform(df['Transported'])
```

```python
cols_drop = ['PassengerId', 'Name','Transported']

df = df.drop(cols_drop, axis = 1)
X = df.values

df.head()
```

```python
df.shape
```

```python
from sklearn.feature_selection import SelectKBest, chi2

selector = SelectKBest(chi2, k=15)

X_best = selector.fit_transform(df, y)

selected_features_mask = selector.get_support()

selected_features = df.columns[selected_features_mask]

print("Fitur terpilih:")
print(selected_features)
```

```python
df.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_best)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
x_latih, x_uji, y_latih, y_uji = train_test_split(X_scaled,y,test_size=0.2, random_state=2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf', 'linear', 'poly', 'sigmoid',]}  

# svm = SVC(C=100, gamma=0.01, kernel='')
svm = SVC()
grid = GridSearchCV(svm, param_grid, refit = True, verbose = 3) 
svm_results = grid.fit(x_latih, y_latih)

# print(grid.score(x_uji,y_uji))

print(svm_results.best_score_)
print(svm_results.best_params_)
```
